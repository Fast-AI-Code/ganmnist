```python
from fastai.vision import *
from fastai.vision.gan import *
```

```python
path = untar_data(URLs.MNIST)
```

```python
def get_data(bs, size):
    return (GANItemList.from_folder(path, noise_sz=100)
               .split_none()
               .label_from_func(noop)
               .transform(tfms=[[crop_pad(size=size, row_pct=(0,1), col_pct=(0,1))], []], size=size, tfm_y=True)
#             .transform(tfms=None, size=size, tfm_y=True)
               .databunch(bs=bs)
               .normalize(stats = [torch.tensor([0.5,0.5,0.5]), torch.tensor([0.5,0.5,0.5])], do_x=False, do_y=True))
```

```python
data = get_data(1024, 32)
```

```python
data.show_batch(rows=5)
```

```python
generator = basic_generator(in_size=32, n_channels=3, n_extra_layers=1).cuda()
critic    = basic_critic   (in_size=32, n_channels=3, n_extra_layers=1).cuda()
```

```python
optim.RMSprop??
```

```python
learn = GANLearner.wgan(data, generator, critic, switch_eval=False,
                        opt_func = partial(optim.RMSprop, lr=0.00005,weight_decay = 0.01))
```

```python
learn.fit(10, 2e-4)
```

```python
learn.save('stage-1')
```

```python
# learn.load('stage-1')
```

```python
learn.unfreeze()
learn.fit(10, 2e-4)
```

```python
learn.save('stage-2')
```

```python
learn.unfreeze()
learn.fit_one_cycle(10)
```

```python
learn.save('stage-3')
```

```python
learn.show_results(ds_type=DatasetType.Train, rows=4, figsize=(16,16))
```

```python
learn.gan_trainer.switch(gen_mode=True)
learn.show_results(ds_type=DatasetType.Train,  rows=4, figsize=(16,16))
```
